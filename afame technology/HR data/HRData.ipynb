## Data cleansing involves removing unnecessary columns.

```python
category_colm = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime', 'Over18']
numerical_colm = []
for i in column_list:
    c = True
    for j in category_colm:
        if i == j:
            c = False
            break 
    if c:
        numerical_colm.append(i)
print(numerical_colm)
```

```python
print(df['EmployeeCount'].value_counts())
print(df['Over18'].value_counts())
print(df['StandardHours'].value_counts())
```

```python
columns_to_drop=['EmployeeCount','Over18','StandardHours','EmployeeNumber','RelationshipSatisfaction']
# all of this column will not effect the prediction of attrition as first three of them are same value for all rows and last two drop colms have not significant impact on column
columns_to_keep = [col for col in df.columns if col not in columns_to_drop]
df = df[columns_to_keep]

df
```
