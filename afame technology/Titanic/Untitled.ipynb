```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
```

```python
df
```

```python
drop_cols = ['Name', 'PassengerId']
df.drop(columns=drop_cols, inplace=True)
```

```python
df.isnull().sum()
```

Null column - Age Cabin

```python
# Assuming df['Gender'] contains 'male' and 'female' values
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
# Create a new column 'Cabin_Missing' indicating if 'Cabin' value is missing
df['Cabin_Missing'] = df['Cabin'].isna()

# Fill missing values in 'Cabin' column with an empty string
df['Cabin'].fillna('', inplace=True)

# Separate 'Cabin' values into 'Cabin_Letter' and 'Cabin_Number'
df[['Cabin_Letter', 'Cabin_Number']] = df['Cabin'].str.extract(r'([A-Za-z]+)(\d+)')

# Convert 'Cabin_Number' column to numeric (optional, depending on further use)
df['Cabin_Number'] = pd.to_numeric(df['Cabin_Number'], errors='coerce')
```

```python
df
```

```python
drop_cols = ['Cabin', 'Ticket']
df.drop(columns=drop_cols, inplace=True)
```

```python
df
```

```python
df['Age'].isnull().sum()
```

```python
skewness = df['Age'].skew()
print("Skewness of Age:", skewness)
```

```python
mean_age = df['Age'].mean()
std_age = df['Age'].std()

# Generate random numbers from a normal distribution
random_ages = np.random.normal(mean_age, std_age, size=df['Age'].isnull().sum())

# Replace NaN values with the generated random numbers
df.loc[df['Age'].isnull(), 'Age'] = random_ages
```

```python
df.isnull().sum()
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
df['Cabin_Letter'].value_counts()
```

```python
import numpy as np

# Assuming df['Cabin_Letter'] contains 'G', 'F', 'E', etc., and NaN for missing values
df['Cabin_Letter'] = df['Cabin_Letter'].map({'G': 0, 'F': 1, 'E': 2, 'D': 3, 'C': 4, 'B': 5, 'A': 6, np.nan: -1})
# Assuming df['Cabin_Missing'] contains True and False for missing values
df['Cabin_Missing'] = df['Cabin_Missing'].map({True: 1, False: 0})
# Assuming df['Cabin_Number'] contains numeric values and NaN for missing values
df['Cabin_Number'] = df['Cabin_Number'].fillna(0)  # Fill NaN with 0, assuming 0 is not a valid value in the original data
```

```python
df
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import  GridSearchCV
rf = RandomForestClassifier()

# Define the hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Use the best model to make predictions
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()

# Define the hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create GridSearchCV
grid_search = GridSearchCV(estimator=gb, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Use the best model to make predictions
best_gb = grid_search.best_estimator_
y_pred = best_gb.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, GradientBoostingClassifier



# Create Gradient Boosting Regressor
gb_regressor = GradientBoostingRegressor()

# Create Random Forest Regressor
rf_regressor = RandomForestRegressor()

# Train the regressors
gb_regressor.fit(X_train, y_train)
rf_regressor.fit(X_train, y_train)

# Get the outputs of the regressors
gb_output = gb_regressor.predict(X_train)
rf_output = rf_regressor.predict(X_train)

# Combine the outputs as input features for the Gradient Boosting Classifier
X_train_ensemble = pd.DataFrame({'GB_Output': gb_output, 'RF_Output': rf_output})

# Create a Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier()

# Define the hyperparameter grid for the classifier
param_grid_classifier = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create GridSearchCV for the classifier
grid_search_classifier = GridSearchCV(estimator=gb_classifier, param_grid=param_grid_classifier, cv=3, n_jobs=-1, verbose=2)

# Fit the grid search to the data
grid_search_classifier.fit(X_train_ensemble, y_train)

# Get the best parameters
best_params_classifier = grid_search_classifier.best_params_
print("Best Parameters for Classifier:", best_params_classifier)

# Use the best model to make predictions
best_gb_classifier = grid_search_classifier.best_estimator_

# Get the test outputs of the regressors
gb_output_test = gb_regressor.predict(X_test)
rf_output_test = rf_regressor.predict(X_test)

# Combine the test outputs as input features for the Gradient Boosting Classifier
X_test_ensemble = pd.DataFrame({'GB_Output': gb_output_test, 'RF_Output': rf_output_test})

# Make predictions using the ensemble classifier
y_pred = best_gb_classifier.predict(X_test_ensemble)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
